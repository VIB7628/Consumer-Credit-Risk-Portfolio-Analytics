```python
import pandas as pd

# Example: Reading a CSV file from your Google Drive
# Make sure to replace 'my_data.csv' with the actual path to your file in Drive
# For example, if your file is in 'My Drive/Colab Notebooks/data.csv', the path would be '/content/drive/My Drive/Colab Notebooks/data.csv'
try:
    df = pd.read_csv('/content/drive/My Drive/my_data.csv')
    print("Data loaded successfully:")
    display(df.head())
except FileNotFoundError:
    print("Error: 'my_data.csv' not found. Please check the file path in your Google Drive.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
df.head(20)
```

```python
df = df.dropna(how='all')
num_cols = ['Credit Score', 'Annual Income']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
df['Years in current job'] = df['Years in current job'].fillna(df['Years in current job'].mode()[0])
zero_fill_cols = ['Months since last delinquent', 'Bankruptcies', 'Tax Liens']
df[zero_fill_cols] = df[zero_fill_cols].fillna(0)
```

```python
df.head(50)
```

```python
print(df.isnull().sum)
```

```python
df.loc[df['Credit Score'] > 850, 'Credit Score'] = df['Credit Score'] / 10
```

```python
import numpy as np
median_loan = df[df['Current Loan Amount'] < 99999999]['Current Loan Amount'].median()
df['Current Loan Amount'] = df['Current Loan Amount'].replace(99999999, median_loan)
```
